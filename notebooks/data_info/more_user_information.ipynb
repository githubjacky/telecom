## the information of the CERT number
- https://zhuanlan.zhihu.com/p/319418575

```python
%%sql 

SELECT CERT_NBR 
FROM tb_asz_cdma_0838_201308
WHERE CERT_NBR RLIKE '[1-8][0-7]-[0-7][0-9]-[0-9][0-9]-[12][09][0-9][0-9]-[01][0-9]-[0-3][0-9]-[01]'
LIMIT 10;
```

- 地址碼: 
    - 51: 四川省

```python
%%sql 

SELECT DISTINCT
    SUBSTR(CERT_NBR, 1, 2) AS area_code,
    COUNT(*)
FROM tb_asz_cdma_0838_201308
WHERE CERT_NBR RLIKE '[1-8][0-7]-[0-7][0-9]-[0-9][0-9]-[12][09][0-9][0-9]-[01][0-9]-[0-3][0-9]-[01]'
GROUP BY area_code
LIMIT 10;
```
